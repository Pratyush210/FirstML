```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
df = pd.read_csv('D:\Dataset\zomato_outlet_final.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(['link'], axis = 1)
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace = True)
df.shape
```

```python
df['dine_rating'].unique()
```

```python
df['dine_rating'].fillna(df['dine_rating'].mean(), inplace = True)
df['dine_rating'].isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace = True)
df.head()
```

```python
df.rename(columns = {'rest_name':'name', 'loc':'location'}, inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['cost'].unique()
```

```python
def handlerate(value):
    
        value = str(value).split('.')
        value = value[1]
        return str(value)
    
df['cost'] = df['cost'].apply(handlerate)
df['cost'].head()
```

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
    
df['cost'] = df['cost'].apply(handlecomma)
df['cost'].unique()
        
```

```python
df.head()
```

```python
df['rest_type'].value_counts()
```

```python
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types
```

```python
rest_types_lessthan100 = rest_types[rest_types<100]
rest_types_lessthan100
```

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan100):
        return 'others'
    else:
        return value
        
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
df['location'].value_counts()
```

```python
location = df['location'].value_counts(ascending  = False)

location_lessthan50 = location[location<50]



def handle_location(value):
    if(value in location_lessthan50):
        return 'others'
    else:
        return value
        
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

```python
df['cuisine'].value_counts()
```

```python
cuisine = df['cuisine'].value_counts(ascending  = False)


cuisine_lessthan20 = cuisine[cuisine<20]



def handle_cuisine(value):
    if(value in cuisine_lessthan20):
        return 'others'
    else:
        return value
        
df['cuisine'] = df['cuisine'].apply(handle_cuisine)
df['cuisine'].value_counts()
```
